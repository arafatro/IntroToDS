```python
x = np.arange(0, math.pi * 2, 0.05)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='blue')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 2, 100)
y = np.exp(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='exp(x)', color='green')
plt.xlabel('x')
plt.ylabel('exp(x)')
plt.title('Exponential Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
coefs = [-2, -1, 0, 1, 2]
x = np.linspace(-2, 2, 100)

fig, axes = plt.subplots(1, 3, sharey=True, figsize=(18, 6))
titles = ['Varying a', 'Varying b', 'Varying c']

for idx, ax in enumerate(axes):
    for coef in coefs:
        if idx == 0:
            y = coef * x**2
            label = f'a={coef}, b=0, c=0'
        elif idx == 1:
            y = x**2 + coef * x
            label = f'a=1, b={coef}, c=0'
        else:
            y = x**2 + coef
            label = f'a=1, b=0, c={coef}'
        ax.plot(x, y, label=label)
        ax.axhline(0, color='black', linewidth=0.5)
        ax.axvline(0, color='black', linewidth=0.5)
        ax.grid(True)
        ax.legend()
        ax.set_title(titles[idx])

plt.suptitle('Quadratic Functions with Varying Coefficients')
plt.show()
```

## Multiple subplots

```python
t = np.arange(0, 5, 0.1)

y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)

plt.subplot(211)
plt.plot(t, y1, 'b-.')
plt.subplot(212)
plt.plot(t, y2, 'r--')
```

```python
t = np.arange(0, 1, 0.1)

y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)
y3 = t

plt.subplot(131)
plt.plot(t, y1, 'b-.')
plt.subplot(132)
plt.plot(t, y2, 'r--')
plt.subplot(133)
plt.plot(t, y3, 'g*')
```

```python
fig, a = plt.subplots(2,2) #subplots(rows, cols)

x = np.arange(0,10)
a[0,0].plot(x, x*x)
a[0,0].set_title('Squared')

a[0,1].plot(x, np.sqrt(x))
a[0,1].set_title('Square root')

a[1,0].plot(x, np.exp(x))
a[1,0].set_title('Exponential')

a[1,1].plot(x, np.log10(x))
a[1,1].set_title('Log10')

plt.show()
```

## Line Plot with Confidence Interval using Seaborn
Seaborn provides a convenient function to plot data with confidence intervals.


```python
# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.1, 100)

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# Plot with Seaborn
sns.lineplot(x='x', y='y', data=data)
plt.title('Line Plot with Confidence Interval')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

##Violin Plot using Seaborn
Violin plots combine aspects of box plots and KDE plots.

```python
# Load the Tips dataset
tips = sns.load_dataset('tips')

# Plot violin plot
sns.violinplot(x='day', y='total_bill', data=tips, palette='muted')
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

##3D Plot using Matplotlib
Matplotlib allows for 3D plotting to visualize data in three dimensions.

```python
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Plot 3D surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title('3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

##Joint Plot using Seaborn
Joint plots display a relationship between two variables along with their marginal distributions.

```python
# Plot joint plot
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', hue='sex', palette='coolwarm')
plt.suptitle('Joint Plot of Total Bill and Tip', y=1.02)
plt.show()
```
