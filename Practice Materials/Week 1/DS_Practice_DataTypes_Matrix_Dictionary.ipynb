## Part 1 — Python Data Types

Before working with DataFrames, it is essential to understand the **basic Python data types**.  
Every value in a dataset maps back to one of these.

| Python Type | Example | Use in DS |
|---|---|---|
| `int` | `42` | Counts, IDs, discrete scores |
| `float` | `3.14` | Measurements, probabilities |
| `str` | `'male'` | Labels, names, categories |
| `bool` | `True` | Flags, binary features |
| `list` | `[1, 2, 3]` | Collections of values |
| `dict` | `{'age': 20}` | Key-value mappings |

````python
# ── 1.1  Basic Python types ──────────────────────────────────────────

age        = 20          # int
gpa        = 14.5        # float
name       = "Alice"     # str
passed     = True        # bool
grades     = [12, 14, 15, 11]   # list
student    = {"name": "Alice", "age": 20, "passed": True}  # dict

print(type(age),    age)
print(type(gpa),    gpa)
print(type(name),   name)
print(type(passed), passed)
print(type(grades), grades)
print(type(student),student)
````

````python
# ── 1.2  Type conversion (casting) ───────────────────────────────────
# Sometimes data arrives in the wrong type and must be converted.

score_as_string = "18"          # read from a CSV as text
print(type(score_as_string))    # str — cannot do maths on this!

score_as_int = int(score_as_string)
print(type(score_as_int), score_as_int + 2)   # now it works

# float → int (truncates decimal)
height = 1.78
print(int(height))      # 1  (not rounded, truncated)
print(round(height))    # 2  (rounded)
````

````python
# ── 1.3  Pandas dtype vs Python type ─────────────────────────────────
# Pandas uses its own type system on top of Python/NumPy.

# Python type  →  Pandas / NumPy dtype
# int          →  int64
# float        →  float64
# str          →  object
# bool         →  bool
# category     →  category  (special Pandas type — memory efficient)

print("Pandas dtypes of the dataset:")
print(df.dtypes)
print("\nNotice: text columns show as 'object' not 'str'")
````

## Part 2 — The Data Matrix

A **data matrix** is the standard rectangular structure used in data science:

- Each **row** = one observation (student, patient, transaction …)
- Each **column** = one feature / variable
- The intersection = one value

```
         col1    col2    col3
row 0  [  v00     v01     v02  ]
row 1  [  v10     v11     v12  ]
row 2  [  v20     v21     v22  ]
```

In Pandas this is a **DataFrame**.

````python
# ── 2.1  Shape of the data matrix ────────────────────────────────────

print("Shape (rows, columns):", df.shape)
print(f"Number of rows    : {df.shape[0]}")
print(f"Number of columns : {df.shape[1]}")
````

````python
# ── 2.2  Exploring the matrix ─────────────────────────────────────────

print("── First 5 rows ──")
display(df.head())

print("── Last 5 rows ──")
display(df.tail())

print("── Random sample of 5 rows ──")
display(df.sample(5, random_state=42))
````

````python
# ── 2.3  Accessing parts of the matrix ───────────────────────────────

# Single column (returns a Series — a 1D slice of the matrix)
print("── Single column: age ──")
print(df["age"].head())

# Multiple columns (returns a DataFrame)
print("\n── Multiple columns ──")
display(df[["age", "studytime", "G1", "G2", "G3"]].head())

# Single row by index
print("\n── Row at index 0 ──")
print(df.iloc[0])

# Specific cell
print("\n── Value at row 0, column 'age' ──")
print(df.at[0, "age"])
````

````python
# ── 2.4  Numerical vs Categorical columns ─────────────────────────────
# Correctly identifying column types is the first step in any analysis.

numerical_cols   = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()

print(f"Numerical columns  ({len(numerical_cols)}):")
print(numerical_cols)

print(f"\nCategorical columns ({len(categorical_cols)}):")
print(categorical_cols)
````

````python
# ── 2.5  Missing values in the matrix ────────────────────────────────
# A missing value (NaN) means no observation was recorded for that cell.

missing_count = df.isnull().sum()
missing_cols  = missing_count[missing_count > 0]

if missing_cols.empty:
    print("No missing values found in this dataset.")
else:
    print("Columns with missing values:")
    print(missing_cols)

# Total missing cells as a percentage
total_cells    = df.shape[0] * df.shape[1]
missing_cells  = df.isnull().sum().sum()
print(f"\nTotal missing cells: {missing_cells} / {total_cells} ({missing_cells/total_cells*100:.2f}%)")
````

````python
# ── 2.6  Basic statistics of the matrix ──────────────────────────────

print("── Descriptive statistics (numerical columns) ──")
display(df.describe())

print("\n── Value counts for a categorical column: 'sex' ──")
print(df["sex"].value_counts())
````

## Part 3 — Data Dictionary

A **data dictionary** is a structured reference table that documents every column in a dataset.  
It is one of the first things you create in a real project — it helps you and your team understand the data.

A good data dictionary contains at minimum:

| Field | Description |
|---|---|
| Column Name | Name as it appears in the dataset |
| Data Type | Pandas dtype (`int64`, `float64`, `object`, …) |
| Missing Values | Number of NaN cells |
| Unique Values | Number of distinct values |
| Description | What the column represents |

````python
# ── 3.1  Build the data dictionary ───────────────────────────────────

# Pull descriptions from the UCI metadata
variable_descriptions = (
    student_performance.variables
    .set_index("name")["description"]
    .to_dict()
)

data_dict = pd.DataFrame({
    "Column Name"  : df.columns,
    "Data Type"    : df.dtypes.astype(str).values,
    "Missing Values": df.isnull().sum().values,
    "Unique Values" : df.nunique().values,
    "Description"  : [variable_descriptions.get(col, "—") for col in df.columns]
})

display(data_dict)
````

````python
# ── 3.2  Enrich the dictionary — add a Category column ───────────────
# We classify each column as Numerical, Categorical, or Binary.

def classify_column(col):
    if df[col].dtype in ["int64", "float64"]:
        if df[col].nunique() == 2:
            return "Binary (Numerical)"
        return "Numerical"
    elif df[col].dtype == "object":
        if df[col].nunique() == 2:
            return "Binary (Categorical)"
        return "Categorical"
    return "Other"

data_dict["Category"] = [classify_column(col) for col in df.columns]

display(data_dict[["Column Name", "Data Type", "Category", "Unique Values", "Description"]])
````

````python
# ── 3.3  Summary of the dictionary ───────────────────────────────────

print("Category breakdown:")
print(data_dict["Category"].value_counts())

print("\nColumns with missing values:")
missing = data_dict[data_dict["Missing Values"] > 0][["Column Name", "Missing Values"]]
print(missing if not missing.empty else "None")
````

````python
# ── 3.4  Type conversion — correcting the data matrix ─────────────────
# Some columns look numerical but are actually categorical.
# Example: 'Medu' (mother education 0-4) is ordinal, not continuous.

print("Before conversion:")
print(df[["school", "sex", "Medu", "Fedu"]].dtypes)

# Convert binary text columns to category
binary_cols = ["school", "sex", "address", "famsize", "Pstatus",
               "schoolsup", "famsup", "paid", "activities",
               "nursery", "higher", "internet", "romantic"]

for col in binary_cols:
    df[col] = df[col].astype("category")

# Convert ordinal columns to category
ordinal_cols = ["Medu", "Fedu", "traveltime", "studytime",
                "famrel", "freetime", "goout", "Dalc", "Walc", "health"]

for col in ordinal_cols:
    df[col] = df[col].astype("category")

print("\nAfter conversion:")
print(df[["school", "sex", "Medu", "Fedu"]].dtypes)
````

````python
# ── 3.5  Rebuild the dictionary after type correction ─────────────────

data_dict_v2 = pd.DataFrame({
    "Column Name"   : df.columns,
    "Data Type"     : df.dtypes.astype(str).values,
    "Missing Values": df.isnull().sum().values,
    "Unique Values" : df.nunique().values,
    "Category"      : [classify_column(col) for col in df.columns],
    "Description"   : [variable_descriptions.get(col, "—") for col in df.columns]
})

print("Updated category breakdown after type correction:")
print(data_dict_v2["Data Type"].value_counts())
display(data_dict_v2)
````

## Part 4 — Putting It All Together: Two Insights
Now that we understand the data types, the matrix, and the dictionary, we extract meaningful insights.

````python
# ── Insight 1: Highest correlation between numerical columns ──────────

num_df   = df.select_dtypes(include=["int64", "float64"])
corr_mat = num_df.corr().abs()

# Remove self-correlations
np.fill_diagonal(corr_mat.values, np.nan)

highest_pair  = corr_mat.stack().idxmax()
highest_value = corr_mat.stack().max()

print(f" Insight 1 — Highest correlation:")
print(f"   '{highest_pair[0]}' and '{highest_pair[1]}' → r = {highest_value:.3f}")
print(f"   Interpretation: These two features are strongly related.")
````

````python
# ── Insight 2: Average final grade by study time ──────────────────────

# Convert studytime back to int for groupby
avg_grade = df.copy()
avg_grade["studytime"] = avg_grade["studytime"].astype(int)

result = avg_grade.groupby("studytime")["G3"].mean().round(2)
result.index = result.index.map({
    1: "<2 hrs",
    2: "2–5 hrs",
    3: "5–10 hrs",
    4: ">10 hrs"
})

print(" Insight 2 — Average Final Grade (G3) by Weekly Study Time:")
print(result.to_string())
print("\n   Interpretation: Students who study more tend to score higher.")
````
